```python
## Data Cleaning:

# Import necessary libraries
import pandas as pd
from google.colab import files

# Upload the dataset
uploaded = files.upload()

# Load the dataset into a Pandas DataFrame
for filename in uploaded.keys():
    df = pd.read_csv(filename)

# Check for missing values in the dataset
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Handle missing values for the 'Financial Stress' column
# Drop rows with missing values in that column
if missing_values['Financial Stress'] > 0:
    df.dropna(subset=['Financial Stress'], inplace=True)

# Check the data types of each column
data_types = df.dtypes
print("Data Types:\n", data_types)

# Convert 'Dietary Habits' to a categorical variable if necessary
df['Dietary Habits'] = df['Dietary Habits'].astype('category')

# Check for duplicates in the dataset
duplicates = df.duplicated().sum()
print("Number of Duplicates:", duplicates)

# Remove duplicates if any exist
if duplicates > 0:
    df.drop_duplicates(inplace=True)

# Provide a summary of the cleaned dataset
cleaned_data_summary = df.describe()
print("Cleaned Data Summary:\n", cleaned_data_summary)
```

```python
# 3. Pie Chart: Proportion of Students Reporting Suicidal Thoughts
yes_count = df['Have you ever had suicidal thoughts ?'].value_counts().get('Yes', 0)
no_count = df['Have you ever had suicidal thoughts ?'].value_counts().get('No', 0)
labels = ['Yes', 'No']
sizes = [yes_count, no_count]
colors = ['lightcoral', 'lightskyblue']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Students Reporting Suicidal Thoughts')
plt.axis('equal')
plt.show()
```

### Insights and Discussion for the Pie Chart: Proportion of Students Reporting Suicidal Thoughts

**Insights:**
- The pie chart depicts the proportion of students reporting suicidal thoughts, with 63.3% indicating "Yes" and 36.7% indicating "No."
- A significant majority of students reported experiencing suicidal thoughts, which highlights a pressing mental health concern within the student population.

**Discussion:**
- The high percentage of students reporting suicidal thoughts underscores the urgent need for mental health resources and support systems within educational institutions.
- Schools and universities should consider implementing comprehensive mental health programs, including counseling services, peer support groups, and awareness campaigns to destigmatize mental health issues.
- Further research is necessary to understand the underlying factors contributing to these feelings, such as academic pressure, social isolation, or personal circumstances.
- Engaging students in discussions about mental health and providing accessible resources can help create a supportive environment and potentially reduce the prevalence of suicidal thoughts.

```python
# 4. Scatter Plot: Study Satisfaction vs. CGPA
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Study Satisfaction', y='CGPA', data=df, color='purple')
plt.title('Study Satisfaction vs. CGPA')
plt.xlabel('Study Satisfaction (1-5)')
plt.ylabel('CGPA')
plt.show()
```

```python
# 5. Heatmap: Correlation Between Numerical Variables
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame with only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Generate the correlation heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
