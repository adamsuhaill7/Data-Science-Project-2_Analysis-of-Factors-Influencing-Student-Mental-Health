```python
# Load the dataset into a Pandas DataFrame
for filename in uploaded.keys():
    df = pd.read_csv(filename)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Handle missing values for the 'Financial Stress' column
if missing_values['Financial Stress'] > 0:
    df.dropna(subset=['Financial Stress'], inplace=True)

# Confirm the action by checking for missing values again
print("Missing Values After Cleaning:\n", df.isnull().sum())
```

```python
# Check the data types of each column
data_types = df.dtypes
print("Data Types:\n", data_types)
```

```python
# Convert 'Dietary Habits' to a categorical variable if necessary
df['Dietary Habits'] = df['Dietary Habits'].astype('category')

# Verify the conversion
print("Data Types After Conversion:\n", df.dtypes)
```

```python
# Check for duplicates in the dataset
duplicates = df.duplicated().sum()
print("Number of Duplicates:", duplicates)
```

```python
# Remove duplicates if any exist
if duplicates > 0:
    df.drop_duplicates(inplace=True)

# Confirm the removal of duplicates
print("Number of Duplicates After Cleaning:", df.duplicated().sum())
```

```python
# Provide a summary of the cleaned dataset
cleaned_data_summary = df.describe()
print("Cleaned Data Summary:\n", cleaned_data_summary)
```

```python
# Import necessary visualisation packages:
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# 1. Bar Plot: Academic Pressure vs. Depression Levels
# Filter the DataFrame to include only Academic Pressure values from 1 to 5
filtered_df = df[df['Academic Pressure'].isin([1, 2, 3, 4, 5])]

plt.figure(figsize=(8, 5))
sns.barplot(x='Academic Pressure', y='Depression', data=filtered_df, palette='viridis')
plt.title('Depression Levels by Academic Pressure')
plt.xlabel('Academic Pressure')
plt.ylabel('Depression (%)')
plt.ylim(0, 1)  # Adjust y-axis limits if needed
plt.show()
```

### Insights and Discussion for the Bar Plot: Depression Levels by Academic Pressure

**Insights:**
- The bar plot shows a clear trend where depression levels (measured as a percentage) increase with higher academic pressure.
- Students experiencing little to no academic pressure (1.0) report the lowest levels of depression, while those under the highest academic pressure (5.0) exhibit the highest depression levels, surpassing 80%.
- The error bars indicate variability within the data, suggesting that while the trend is clear, individual experiences of depression can vary significantly even within the same academic pressure category.

**Discussion:**
- The increasing trend in depression levels with academic pressure highlights a concerning relationship that warrants attention. Research by *Beiter et al. (2015)* indicates that academic stress is a significant factor contributing to mental health issues among students, supporting the need for institutions to implement mental health support programs to assist students managing high levels of academic stress.
- This visualization prompts further investigation into the underlying factors contributing to both academic pressure and depression. Are there specific aspects of academic life (e.g., workload, expectations) that are particularly detrimental to mental health? A study by *Hudd et al. (2000)* identifies specific stressors, including workload and time management, as critical elements affecting students' mental well-being.
- Additionally, qualitative studies could complement this quantitative data, providing deeper insights into students' experiences and coping mechanisms. For instance, *O’Connor et al. (2020)* emphasize the importance of understanding personal narratives in addressing mental health issues.
- Addressing academic pressure through workshops, counseling, and stress management programs could mitigate its negative effects on student mental health. Evidence from *Conley et al. (2013)* suggests that such interventions can lead to improved mental health outcomes.

```python
# 2. Stacked Bar Plot: Financial Stress vs. Depression Levels
# Filter the DataFrame to include only Financial Stress values from 1 to 5
filtered_df = df[df['Financial Stress'].isin([1, 2, 3, 4, 5])]

# Calculate the counts for each combination of Financial Stress and Depression
count_df = filtered_df.groupby(['Financial Stress', 'Depression']).size().reset_index(name='Count')

# Pivot the DataFrame to get a suitable format for plotting
pivot_df = count_df.pivot(index='Financial Stress', columns='Depression', values='Count').fillna(0)

# Normalize to get proportions
pivot_df = pivot_df.div(pivot_df.sum(axis=1), axis=0)

plt.figure(figsize=(8, 5))
pivot_df.plot(kind='bar', stacked=True, color=['#00bfc4', '#f8766d'], alpha=0.8, ax=plt.gca())
plt.title('Depression Levels by Financial Stress')
plt.xlabel('Financial Stress')
plt.ylabel('Proportion')
plt.ylim(0, 1)  # Ensure y-axis goes from 0 to 1
plt.xticks(rotation=0)
plt.legend(title='Depression', labels=['0 (No)', '1 (Yes)'])
plt.show()
```

### Insights and Discussion for the Stacked Bar Plot: Depression Levels by Financial Stress

**Insights:**
- The stacked bar plot illustrates the proportions of students experiencing different levels of financial stress (ranging from 1 to 5) alongside their reported depression levels (0 for "No" and 1 for "Yes").
- As financial stress increases, the proportion of students reporting depression also tends to rise, indicating a potential correlation between financial challenges and mental health issues.

**Discussion:**
- The observed trend where higher levels of financial stress correlate with increased reports of depression underscores the significant mental health challenges faced by students. Research by *Huang et al. (2020)* highlights that financial stress is a strong predictor of anxiety and depression among college students, suggesting that financial pressures exacerbate mental health issues.
- Educational institutions should consider implementing targeted mental health support systems that specifically address the needs of financially stressed students. This could include access to counseling services, workshops on financial management, and peer support programs, as recommended by *Eisenberg et al. (2009)*.
- Understanding the underlying factors contributing to this relationship is crucial. For example, academic pressure, job insecurity, or lack of financial literacy may play significant roles in how financial stress impacts mental health. A study by *Jiang et al. (2022)* discusses how financial literacy can mitigate the adverse effects of financial stress.
- Creating open dialogues about mental health and financial stress can help destigmatize these issues, encouraging students to seek help without fear of judgment. *Baker et al. (2017)* emphasize the importance of communication in reducing stigma associated with mental health.
- Future research is necessary to evaluate the effectiveness of interventions designed to alleviate financial stress and their impact on mental health outcomes.
```python
# 3. Pie Chart: Proportion of Students Reporting Suicidal Thoughts
yes_count = df['Have you ever had suicidal thoughts ?'].value_counts().get('Yes', 0)
no_count = df['Have you ever had suicidal thoughts ?'].value_counts().get('No', 0)
labels = ['Yes', 'No']
sizes = [yes_count, no_count]
colors = ['lightcoral', 'lightskyblue']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Students Reporting Suicidal Thoughts')
plt.axis('equal')
plt.show()
```

```python
# 5. Heatmap: Correlation Between Numerical Variables
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame with only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Generate the correlation heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

### Insights and Discussion for the Correlation Heatmap

**Insights:**
- The heatmap displays the correlation coefficients between various factors, including Age, Academic Pressure, Work Pressure, CGPA, Study Satisfaction, Job Satisfaction, Work/Study Hours, Financial Stress, and Depression.
- Interesting correlations include:
  - **Academic Pressure and Depression (0.47)**: Higher academic pressure is associated with increased levels of depression.
  - **Financial Stress and Depression (0.36)**: Increased financial stress correlates with higher levels of depression.
  - **Work Pressure and Job Satisfaction (0.77)**: Higher work pressure is strongly associated with higher job satisfaction.
  - **Study Satisfaction and Depression (-0.17)**: Higher study satisfaction is linked to lower levels of depression.
  - **Study Hours and Depression (0.21)**: Increased study hours are associated with higher levels of depression.

**Discussion:**
- The correlation between academic pressure and depression suggests that managing academic workloads could be vital for improving student mental health. Research by *Beiter et al. (2015)* emphasizes the importance of balancing academic demands to mitigate stress.
- The link between financial stress and depression indicates a need for support systems that address financial challenges faced by students. Studies, such as those by *Huang et al. (2020)*, highlight the mental health implications of financial insecurity.
- The strong positive correlation between work pressure and job satisfaction suggests that some individuals may find high pressure motivating, highlighting the need for understanding individual differences in response to work demands. *Hackman and Oldham (1976)* discuss how certain job characteristics can enhance satisfaction despite pressure.
- The negative correlation between study satisfaction and depression implies that fostering a more satisfying study experience may help reduce depressive symptoms. According to *Hudd et al. (2000)*, enhancing student engagement can lead to improved mental health outcomes.
- Additionally, the positive correlation between study hours and depression suggests that excessive studying could contribute to mental health issues, warranting a balanced approach to study time. Research by *Misra and McKean (2000)* indicates that time management is crucial for student well-being.
- Understanding these interrelationships can guide educational institutions in developing targeted interventions, such as stress management programs and financial counseling, to improve both mental health and academic outcomes. *Conley et al. (2013)* advocate for comprehensive mental health services in academic settings.
- Further research could explore the causal relationships within these correlations, examining how targeted interventions might positively affect multiple areas of student life. Longitudinal studies, as suggested by *Astin (1993)*, could provide valuable insights into these dynamics.
